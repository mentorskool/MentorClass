* Login to your Github account [here](https://github.com/login)
* Create a new repository named : **lab-first-git**
  * Give a nice one line **description**
  * Choose **public**
  * Select **Initialize this repository with a README**
* Clone this repo (anywhere) in your local machine
* Add the file **sample.md** downloaded in Step-2 above into this file
* Check **git status**
* Do **git add**
* Perform **commit**
* **Push** the file to **remote** git location
* Switch to your git repo on your GitHub account and check if it reflects the newly added file or not
* Switch back to your local folder (which has the file)
  * Check for new folder getting created (called .git)
