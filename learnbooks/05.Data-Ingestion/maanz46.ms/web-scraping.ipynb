## Import Libraries

```
import requests #import requests
from bs4 import BeautifulSoup #import beautifulsoup
```

##Pass HTTP request to server

```
url = 'https://en.wikipedia.org/wiki/List_of_largest_manufacturing_companies_by_revenue'
response = requests.get(url)
response.status_code
```

## Content of the request in html structure

```
soup = BeautifulSoup(response.content,'html.parser')
print(soup.prettify())
```

```
type(soup) 
```

##Extracting information from the web

```
table = soup.find('table',class_ = 'wikitable sortable plainrowheads').tbody
print(table)
```

```
rows = table.find_all('tr') 
print(rows)
```

```
rows[0].find_all('th')
```

```
[v.text for v in rows[0].find_all('th')]
```

```
columns = [v.text.replace('\n','') for v in rows[0].find_all('th')]
print(columns)
```

```
import pandas as pd
df = pd.DataFrame(columns=columns)
```

```
for i in range(1,len(rows)):
  tds = rows[i].find_all('td')
  values = [td.text.replace('\n','').replace('\xa0','').replace('x','') for td in tds]
  df = df.append(pd.Series(values,index=columns),ignore_index=True)
df
```

```
base_url = 'https://en.wikipedia.org'
for i in rows[1:]:
 print(base_url + i.find('a').get('href'))
```
