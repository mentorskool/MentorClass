## Import Libraries

```
import pandas as pd
import json
import requests
import math 
```

### OPEN API 

#### Openaq

```
city_api ='https://api.openaq.org/v1/cities'               #Provides a simple listing of cities within the platform.
country_api = 'https://api.openaq.org/v1/countries'        #Provides a simple listing of countries within the platform.
latest_api ='https://api.openaq.org/v1/latest'             #Provides the latest value of each available parameter for every location in the system.
location_api = 'https://api.openaq.org/v1/locations'       #Provides a list of measurement locations and their meta data.
measurement_api = 'https://api.openaq.org/v1/measurements' #Provides data about individual measurements
```

##### Accessing the data in city_api

```
source = requests.get(city_api)
data = source.text
data
```

##### Converting json data into dictionary

```
data_dict = json.loads(data)
data_dict
```

##### Accessing the meta part from dictionary

```
meta_data = data_dict['meta']
meta_data
```

##### Calculating the number of pages in json

```
count = math.ceil(meta_data['found']/meta_data['limit'])
count
```

##### Extracting the data for listed cities in India

```
final_data = pd.DataFrame()
for i in range(1,31):
    source = requests.get(f"{'https://api.openaq.org/v1/cities'}?page={i}")
    data = source.text 
    data_dict = json.loads(data)  
    # print(data_dict)                          
    pollution_data= data_dict['results']
    #print(pollution_data)
    df_pollution = pd.DataFrame(pollution_data)
    if(True in set(df_pollution['country'] == 'IN')):    #Checking wheher the IN is present in the country column of df_pollution DataFrame
      final_data=final_data.append(df_pollution[df_pollution['country'] == 'IN'],ignore_index=True)
```

```
final_data.head()
```

```
final_data.to_csv('final_data.csv')
```
