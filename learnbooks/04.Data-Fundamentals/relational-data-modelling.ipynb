### **What do we mean by Data Modelling?**
--------------------------------
* It is the **art of transforming** a **real world** working of a **business** into a set of **inter-related tables** in a **Database**
* It is **needed** by every **web/mobile application** built on earth because without a well modelled database, these applications will simply dump **un-related**, **out-of-order** data which would be of no **real business utility**
* Data Modelling is an **attempt to emulate** what's going on in the business in the **real world**. Ex : 
  * a **Customer** registering on the website
  * an **Order** being placed by a Customer
  * a **Product** being added into the shopping cart as part of an **Order**
  * a **Vendor** delivering the Order to the **Customer**


### **What are entities in a business?**
--------------------------------
An **entity** is a **real-world object** **related to** a **business** or otherwise that is **represented in database**. It can be any **object,place,person or class**.

In **DBMS** we store data in the form of **table** containing information about** entity type** like **students,teachers,employees** etc

### **What is Normalization and why is it needed?**
--------------------------------
* **Problem of Data Redundancy**
  * Redundancy means repetition of data many times in a table
  * Redundancy was a big problem in the early stages of database development (during 1970s) when storage space was a big luxury!
